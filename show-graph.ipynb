```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

file_path = '/content/drive/MyDrive/bert_result.xlsx'

start_row = 138
end_row = 147

df = pd.read_excel(file_path,
                   names = ['epoch', 'batch_size', 'learning_rate', 'train_loss', 'train_accuracy', 'val_loss', 'val_accuracy'],
                   usecols = [1, 2, 3, 4, 5, 6, 7])
df = df.iloc[start_row-2:end_row-1, :]
```

```python
df
```

```python
fig, loss_ax = plt.subplots()
acc_ax = loss_ax.twinx()

loss_ax.plot(df['epoch'], df['train_loss'], 'y', label = 'train loss')
loss_ax.plot(df['epoch'], df['val_loss'], 'r', label = 'val loss')
acc_ax.plot(df['epoch'], df['train_accuracy'], 'b', label = 'train accuracy')
acc_ax.plot(df['epoch'], df['val_accuracy'], 'g', label = 'val accuray')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuracy')

line_pos = 3

plt.axvline(x=line_pos, ymin=0.15, ymax=0.85, color='k', linestyle=':')

loss_ax.legend(loc = 'upper left')
acc_ax.legend(loc = 'lower left')

val1 = df[df['epoch'] == line_pos]['train_loss'].values[0]
val2 = df[df['epoch'] == line_pos]['val_loss'].values[0]
val3 = df[df['epoch'] == line_pos]['train_accuracy'].values[0]
val4 = df[df['epoch'] == line_pos]['val_accuracy'].values[0]

xpos = 0.38
ypos = 0.93
loss_ax.text(xpos, ypos, f'train loss: {val1}', transform=loss_ax.transAxes)
loss_ax.text(xpos, ypos-0.05, f'val loss: {val2}', transform=loss_ax.transAxes)
acc_ax.text(xpos, ypos-0.1, f'train accuracy: {val3}', transform=acc_ax.transAxes)
acc_ax.text(xpos, ypos-0.15, f'val accuracy: {val4}', transform=acc_ax.transAxes)

plt.suptitle('batch_size=128, learning_rate=2e-5')

plt.savefig('/content/drive/MyDrive/graphs/graph-bs-128-lr-2e5')

plt.show()
```
